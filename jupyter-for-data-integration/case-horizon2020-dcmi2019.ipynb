```python
import csv
import collections
import xml.etree.ElementTree as ET
import json
```

```python
def head(filename, lines=10, encoding='utf-8', maxcols=None):
    with open(filename, 'r', encoding=encoding) as f:
        for i in range(lines):
            line = f.readline().rstrip('\n');
            if maxcols:
                line = line[:maxcols]
            print(line)
```

```python
head('../data/cordis-h2020projects.csv', 1)
```

```python
with open('../data/cordis-h2020projects.csv') as f:
    reader = csv.DictReader(f, delimiter=';')
    overall = 0
    coordinator_countries = collections.Counter()
    for row in reader:
        overall += 1
        coordinator_countries[row['coordinatorCountry']] += 1
print('Overall projects: {}'.format(overall))
print('Coordinators per country:',coordinator_countries)
        
```

```python
tree = ET.parse('../data/API_SP.POP.TOTL_DS2_en_xml_v2_151522.xml')
root = tree.getroot()
data = root[0]

population = {}

for child in data:
    if child.find("field[@name='Year']").text == '2018':
        ccode3 = child.find("field[@name='Country or Area']").get('key')
        pop = child.find("field[@name='Value']").text
        if pop == None:
            print("Country {} has no population".format(ccode3))
            continue
        population[ccode3] = int(pop)
print("Population of Great Britain:", population['GBR'])
```
